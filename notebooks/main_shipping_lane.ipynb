```python
import xarray as xr
import matplotlib.pyplot as plt
import proplot as plot
```

## [Global traffic density data](https://datacatalog.worldbank.org/dataset/global-shipping-traffic-density)

Using the data to check "straight" shipping lanes

```python
da = xr.open_rasterio('../data/shipdensity_global/shipdensity_global.tif', chunks=(1, 1e4, 1e4))
```

```python
print(da)
```

Because the data is quite large, I cheacked the density online: https://www.arcgis.com/home/webmap/viewer.html?webmap=7838f608a0214d968a118f85d48efd93.
Then, pick the shipping lanes in detail below:

### Indian Ocean

```python
subset_da = da.sel(x=slice(70, 105), y=slice(20, -10)).isel(band=0)
       
fig, ax = plot.subplots(proj='pcarree', figsize=(8, 6))
m = ax.pcolormesh(subset_da.x, subset_da.y, subset_da, cmap='magma', vmin=0, vmax=2e7, extend='max')

ax.format(title='Eastern Indian Ocean',
          lonlim=(70, 105),
          latlim=(-10, 20),
          coast=True,
          labels=True,
          reso='med',
          coastcolor='w',
          gridcolor='w',
          gridlabelcolor='k',
          gridlinestyle='--',
          gridlinewidth=2,
          dms=False)

fig.colorbar(m, loc='b')
```
